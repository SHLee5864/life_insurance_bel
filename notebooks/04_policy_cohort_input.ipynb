```python
import pandas as pd

VERSION_ID = "COHORT_2026_04"

cohort_data = [
    ("C1", "M", 35, 30, 65, 100000, 3200),
    ("C1", "F", 35, 30, 65, 100000, 2400),
    ("C2", "M", 50, 25, 75, 100000, 2600),
    ("C2", "F", 50, 25, 75, 100000, 2200),
    ("C3", "M", 60, 20, 80, 100000, 1600),
    ("C3", "F", 60, 20, 80, 100000, 1700),
    ("C4", "M", 65, 15, 80, 100000, 900),
    ("C4", "F", 65, 15, 80, 100000, 1300),
]

rows = []
for cohort_id, sex, issue_age, term_years, maturity_age, sum_assured, policy_count in cohort_data:
    valuation_age = maturity_age - term_years
    elapsed = valuation_age - issue_age  # 설계서 기반 계산
    rows.append({
        "cohort_id": cohort_id,
        "sex": sex,
        "issue_age": issue_age,
        "term_years": term_years,
        "maturity_age": maturity_age,
        "sum_assured": sum_assured,
        "policy_count": policy_count,
        "elapsed_duration_years": elapsed,
        "version_id": VERSION_ID
    })

df = pd.DataFrame(rows)
```

```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()

df_spark = spark.createDataFrame(df)

(
    df_spark
    .write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("workspace.life_insurance_raw.policy_cohort_input")
)

display(df_spark)
```
