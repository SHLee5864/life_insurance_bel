```python
# parameters
VERSION_ID = "LAPSE_2026_04"
MAX_DURATION = 30

def lapse_rate_for_duration(d: int) -> float:
    if d == 1:
        return 0.15
    elif d == 2:
        return 0.13
    elif d == 3:
        return 0.11
    elif d == 4:
        return 0.09
    elif d == 5:
        return 0.08
    elif 6 <= d <= 10:
        return 0.06
    elif 11 <= d <= 20:
        return 0.03
    else:  # 21–30
        return 0.02

def generate_lapse_curve(max_duration: int = 30):
    return [
        (d, lapse_rate_for_duration(d))
        for d in range(1, max_duration + 1)
    ]

lapse_data = generate_lapse_curve(MAX_DURATION)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import lit

spark = SparkSession.builder.getOrCreate()

df_lapse = (
    spark.createDataFrame(lapse_data, ["duration_year", "lapse_rate_annual"])
    .withColumn("version_id", lit(VERSION_ID))
)

(
    df_lapse
    .write
    .format("delta")
    .mode("overwrite")
    .saveAsTable("workspace.life_insurance_raw.lapse_assumption")
)

display(df_lapse)
```
